```python
driver = webdriver.Chrome()
```

```python
wait = WebDriverWait(driver, 2)
```

```python
campo = wait.until(
   EC.presence_of_all_elements_located(
       (By.ID, "twotabsearchtextbox")
   )
)

campo
```
